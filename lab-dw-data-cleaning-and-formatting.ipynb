## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# --- Columna: 'gender' ---
print("Valores únicos 'gender' ANTES:", df['gender'].unique())

gender_mapping = {
    'F': 'Female',
    'Femal': 'Female',
    'Female': 'Female',
    'M': 'Male',
    'Male': 'Male'
}
df['gender'] = df['gender'].replace(gender_mapping)
print("Valores únicos 'gender' DESPUÉS:", df['gender'].unique())
```

```python
# --- Columna: 'state' ---
print("Valores únicos 'state' ANTES:", df['state'].unique())

state_mapping = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'California': 'California',
    'Arizona': 'Arizona',
    'WA': 'Washington',
    'Washington': 'Washington',
    'NV': 'Nevada',
    'Nevada': 'Nevada',
    'OR': 'Oregon',
    'Oregon': 'Oregon'
}
df['state'] = df['state'].replace(state_mapping)
print("Valores únicos 'state' DESPUÉS:", df['state'].unique())
print("-" * 50)
```

```python
df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})

df['vehicle_class'] = df['vehicle_class'].replace({'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'})

df.head()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '', regex=False)

df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

df.head()
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python

import numpy as np
import pandas as pd


def extraer_valor_medio_estricto(val):
    # Si ya es nulo, lo mantenemos como tal
    if pd.isna(val):
        return np.nan

    val_str = str(val).strip()

    # Nos ceñimos estrictamente al patrón con '/' solicitado
    if "/" in val_str:
        partes = val_str.split("/")
        # Filtramos posibles espacios vacíos
        partes = [p.strip() for p in partes if p.strip() != ""]

        # Para un formato tipo '1/5/00', la lista tiene 3 elementos; el medio es el índice 1
        if len(partes) >= 3:
            return partes[1]
        elif len(partes) == 2:
            return partes[
                0
            ]  # Manejo seguro si solo viniera un separador (ej. '1/5')
        elif len(partes) == 1:
            return partes[0]

    # Si no cumple el patrón del ejercicio, devolvemos NaN para evitar inconsistencias de texto
    return np.nan


# Aseguramos el nombre estandarizado de la columna (con guiones bajos)
df["number_of_open_complaints"] = df["number_of_open_complaints"].apply(
    extraer_valor_medio_estricto
)

# Convertimos a numérico de forma segura
df["number_of_open_complaints"] = pd.to_numeric(
    df["number_of_open_complaints"], errors="coerce"
)
```

```python
# Ahora la lista es 100% consistente con el resto del notebook
columnas_num = [
    'customer_lifetime_value', 
    'income', 
    'monthly_premium_auto', 
    'number_of_open_complaints', 
    'total_claim_amount'
]

# Imputamos los nulos con la mediana de forma segura
for col in columnas_num:
    if col in df.columns:
        # Aseguramos que sea numérico antes de calcular la mediana
        df[col] = pd.to_numeric(df[col], errors='coerce')
        mediana = df[col].median()
        df[col] = df[col].fillna(mediana)
```

```python
# Buscamos la moda (el valor más común, ej. 'Female')
moda_gender = df['gender'].mode()[0]

# Rellenamos los nulos de forma persistente
df['gender'] = df['gender'].fillna(moda_gender)

# Comprobamos directamente
print(df.isnull().sum())
```

```python
# ==============================================================================
# PASO 1: Identificación y exploración de filas duplicadas
# ==============================================================================
# Usamos .duplicated() para contar cuántas filas exactas están repetidas
total_duplicados = df.duplicated().sum()
print(f"Análisis inicial: Se encontraron {total_duplicados} filas completamente duplicadas.")

# ==============================================================================
# PASO 2: Estrategia, eliminación persistente y reseteo de índice
# ==============================================================================
"""
DOCUMENTACIÓN Y JUSTIFICACIÓN:
Estrategia: Eliminación completa de registros duplicados manteniendo la primera ocurrencia (keep='first').
Razonamiento: Cada fila representa una entidad única (cliente/póliza). Las filas idénticas en todas 
sus columnas son anomalías (errores de duplicación en la carga de datos). Mantenerlas sesgaría 
los análisis estadísticos (como medias y varianzas) al inflar artificialmente ciertos perfiles.
"""

# Eliminamos duplicados y REASIGNAMOS a 'df' para que el cambio sea persistente
df = df.drop_duplicates(keep='first')

# Reseteamos el índice para mantener la consistencia secuencial (0, 1, 2...) tras borrar filas
# 'drop=True' evita que el índice desordenado viejo se convierta en una columna nueva
df = df.reset_index(drop=True)
print("-> Duplicados eliminados e índice reseteado con éxito.")

# ==============================================================================
# PASO 3: Verificación del éxito del proceso
# ==============================================================================
# Comprobamos de nuevo con .duplicated() para asegurar que el contador baje a 0
duplicados_post = df.duplicated().sum()
print(f"Verificación final: Filas duplicadas restantes = {duplicados_post}")

# ==============================================================================
# PASO 4: Guardado del dataset limpio en un nuevo archivo CSV
# ==============================================================================
# Exportamos a CSV usando index=False para no añadir una columna de índices innecesaria
archivo_salida = 'customer_analysis_clean.csv'
df.to_csv(archivo_salida, index=False)
print(f"-> Archivo de cierre exportado correctamente como: '{archivo_salida}'")

# Visualización de control
df.head()
```

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting.

```python
import pandas as pd

# ==============================================================================
# 1. ASEGURAR TIPOS DE DATOS CORRECTOS
# ==============================================================================
# Nos aseguramos de que las columnas críticas sean numéricas antes de los cuantiles
columnas_analisis = ["income", "customer_lifetime_value"]

for col in columnas_analisis:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Eliminamos filas que hayan quedado con NaN en estas columnas para no falsear los percentiles
df_base_limpia = df.dropna(subset=columnas_analisis)

# ==============================================================================
# 2. CÁLCULO DE UMBRALES (PERCENTILES)
# ==============================================================================
# Definimos los percentiles con nombres claros y específicos
UMBRAL_INCOME_PERCENTIL_25 = df_base_limpia["income"].quantile(0.25)
UMBRAL_CLV_PERCENTIL_75 = df_base_limpia["customer_lifetime_value"].quantile(
    0.75
)

print("--- Umbrales Calculados para el Filtro ---")
print(f"Percentil 25 de Income (Mínimo): {UMBRAL_INCOME_PERCENTIL_25}")
print(f"Percentil 75 de CLV (Mínimo): {UMBRAL_CLV_PERCENTIL_75}\n")

# ==============================================================================
# 3. FILTRADO SIMULTÁNEO Y SELECCIÓN DE COLUMNAS
# ==============================================================================
# Aplicamos ambas condiciones a la vez (Filtro Cruzado)
condicion_income = df_base_limpia["income"] >= UMBRAL_INCOME_PERCENTIL_25
condicion_clv = (
    df_base_limpia["customer_lifetime_value"] >= UMBRAL_CLV_PERCENTIL_75
)

# Creamos un DataFrame con un nombre descriptivo que contiene solo el subconjunto filtrado
df_clientes_alto_valor = df_base_limpia[
    condicion_income & condicion_clv
].copy()

# ==============================================================================
# 4. VISUALIZACIÓN Y DESCRIPCIÓN ESTADÍSTICA
# ==============================================================================
print("--- RESUMEN ESTADÍSTICO DEL SUBCONJUNTO FILTRADO ---")
# Seleccionamos ambas columnas simultáneamente y aplicamos .describe()
estadisticas_subconjunto = df_clientes_alto_valor[columnas_analisis].describe()
print(estadisticas_subconjunto)

print("\n--- VISTA PREVIA DEL DATAFRAME FILTRADO (Inspección Clara) ---")
# Mostramos las primeras filas del subconjunto centrándonos en las columnas clave
df_clientes_alto_valor[columnas_analisis].head(10)
```
